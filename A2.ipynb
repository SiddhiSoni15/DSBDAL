```python
#detecting outliers using inter quantile range(IQR)

Q1=df['Percentage'].quantile(0.25)
Q3=df['Percentage'].quantile(0.75)
IQR=Q3-Q1

lower_limit=Q1-1.5*IQR
upper_limit=Q1+1.5*IQR

print(f'Q1={Q1},\nQ3={Q3},\nIQR={IQR},\nlower_limit={lower_limit},\nupper_limit={upper_limit}')
```

```python
df[(df['Percentage']<lower_limit)|(df['Percentage']>upper_limit)]
```

```python
#detect outlier using z-score
from scipy import stats

def detect_outliers_zscore(df,threshold=2):
    outliers={}
    for col in df.select_dtypes(include=['int64','float64']).columns:
        z_scores=np.abs(stats.zscore(df[col]))
        outliers[col]=df[col][z_scores>threshold].values
    return outliers

outliers_found=detect_outliers_zscore(df)
for column,outlier_values in outliers_found.items():
    print(f"Outliers in '{column}':{outlier_values if len(outlier_values)>0 else 'None'}")
```

```python
def calculate_zscore(column):
    mean = column.mean()  
    std_dev = column.std()  
    return (column - mean) / std_dev  


z_scores = df.select_dtypes(include=['int64', 'float64']).apply(calculate_zscore)


threshold = 2
outliers = (np.abs(z_scores) > threshold)


for column in df.select_dtypes(include=['int64', 'float64']).columns:
    outlier_values = df[column][outliers[column]]
    print(f"Outliers in '{column}': {outlier_values.tolist() if not outlier_values.empty else 'None'}")
```

```python
#detect outlier using scatter plot
for column in df.select_dtypes(include=['int64','float64']).columns:
    plt.figure(figsize=(4,2))
    plt.scatter(df.index,df[column],color='blue',alpha=0.3)
    plt.xlabel("Index")
    plt.ylabel(column)
    plt.title(f"Scatter Plot of {column}")
    plt.show()
                                       
                                       
                                       
```

```python
#removing outliers

outliers=[]
for i in df.Percentage:
    if i<lower_limit or i>upper_limit:
        outliers.append(i)
print("Outliers are ",outliers)
```
